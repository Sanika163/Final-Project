```python
# import library
import pandas as pd
import seaborn
```

```python
# import data
attrition = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/EmployeeAttrition.csv')
     
```

```python
# view data
attrition.head()
```

```python
# info of data
attrition.info()
```

```python
# summary statistics
attrition.describe()
```

```python
# check for missing value
attrition.isna().sum()
```

```python
# check for categories
attrition.nunique()
```

```python
# visualize pairplot
seaborn.pairplot(attrition)
```

```python
# columns name
attrition.columns
```

```python
# column names of numerical columns
import numpy as np
attrition.select_dtypes(include=np.number).columns
```

```python
# define y
y=attrition['Attrition']
```

```python
# define X
X = attrition.select_dtypes(include=np.number)
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
     
```

```python
# verify shape
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
```

```python
# train model
rfc= RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
# predict with model
y_pred=rfc.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
     
```

```python
# model accuracy
accuracy_score(y_pred,y_test)
```

```python
# model confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
# model classification report
print(classification_report(y_test, y_pred))
```
